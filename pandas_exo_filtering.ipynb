## Filtering and sorting data - chipotle

```python
import pandas as pd
chipo=pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")
```

```python
chipo.head(5)
```

```python
chipo['item_price']=chipo['item_price'].str.replace("$","")
```

```python
chipo['item_price'] = chipo['item_price'].astype(float)
```

```python
chipo.head()
```

How many products cost more than $10.00?

```python
df=chipo.loc[chipo['item_price'] >10]
df['item_name'].count()
```

What is the price of each item?

```python
df.filter(items=['item_name','item_price'])
```

```python
df2=chipo.query('item_price>10')
```

```python
df2.drop_duplicates()
```

```python
df2.head()
```

Sort by the name of the item

```python
df2.sort_values('item_name')
```

```python
chipo.sort_values(by='item_name')
```

```python
chipo['item_price'].max()
```

```python
chipo['item_price'].idxmax()
```

```python
dmax
```
