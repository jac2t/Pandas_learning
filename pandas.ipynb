```python
import pandas as pd

df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/6_handling_missing_data_replace/weather_data.csv")
df
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv",
               parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)
```

```python
df
```

```python
## we can replace all NaN by a value using fillna
new_df=df.fillna(0)
new_df
```

```python
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':"no event"
})
new_df
```

```python
## filling with 0 does not really make sense. we can carry forward the values from previous days
```

```python
new_df = df.fillna(method="ffill") ## ffill for "forward fill"
new_df
```

```python
new_df = df.fillna(method="bfill") ## bfill for "back fill" 
new_df
```

```python
## check fillna documentation for limit and other methods
```

Dealing with missing values using "Interpolate"

```python
new_df=df.interpolate()
new_df
```

```python
new_df=df.interpolate(method="time") ## see result, it takes into account the distance between avaible date and missing date
new_df
```

```python
new_df=df.dropna(how='all')
new_df
```

```python
new_df=df.dropna(thresh=1) # here we keep rows that have at least one valid value
new_df
```

```python
new_df=df.dropna(thresh=2) # here we keep rows that have at least two valid value
new_df
```

concatenate many data frames

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df=pd.concat([india_weather,us_weather])
df
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True)
df
```

Adding keys to concat

```python
df = pd.concat([india_weather, us_weather], keys=["india", "us"])
df
```

```python
df.loc['us']
```

#Merge

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3=pd.merge(df1,df2,on='city') #just like join in sql
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando","baltimore"],
    "temperature": [21,14,35,45],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","San francisco"],
    "humidity": [65,68,75],
})
df2
```

```python
df3=pd.merge(df1,df2,on='city') 
df3
```

```python
## here we're geting only the elements that are common between the two DF
```

```python
df3=pd.merge(df1,df2,on='city',how='outer') ##using the union of DFs. outer join in sql
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='left') # left join
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='right') # right join
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='outer', indicator=True)
df3
```
